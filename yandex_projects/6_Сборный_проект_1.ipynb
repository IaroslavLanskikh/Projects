````python
ratio_list = []
for row in data.groupby('platform'):
    row_row = row[1]
    total = len(row_row)
    
    nan_count_critic_score = len(row_row.query('critic_score.isna()'))
    nan_count_user_score = len(row_row.query('user_score.isna()'))
    nan_count_rating = len(row_row.query('rating.isna()'))
    
    ratio_critic_score = nan_count_critic_score / total
    ratio_user_score = nan_count_user_score / total
    ratio_rating = nan_count_rating / total
    
    ratio_list.append({'platform':row[0],'ratio_critic_score':ratio_critic_score,'ratio_user_score':ratio_user_score, 'ratio_rating':ratio_rating})
````

    –í–∏–¥–Ω–æ, –∫–∞–∫–æ–π –æ–≥—Ä–æ–º–Ω—ã–π –ø—Ä–æ—Ü–µ–Ω—Ç –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Ä–µ–≥–∏–æ–Ω–µ JP. –•–æ—Ç—è —É –æ—Å—Ç–∞–ª—å–Ω—ã—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤ —Ç–æ–∂–µ –±–æ–ª—å—à–æ–π –ø—Ä–æ—Ü–µ–Ω—Ç, –±–æ–ª—å—à–µ 30%.

üëç –í–µ—Ä–æ—è—Ç–Ω–æ –∏—Å—Ç–æ—á–Ω–∏–∫ –º–∞–ª–æ –∑–Ω–∞–µ—Ç –ø—Ä–æ —è–ø–æ–Ω—Å–∫–∏–µ –∏–≥—Ä—ã. –¢–∞–º, –Ω–∞–ø—Ä–∏–º–µ—Ä, –µ—Å—Ç—å —Å–≤–æ–π –∞–Ω–∞–ª–æ–≥ ESRB —Ä–µ–π—Ç–∏–Ω–≥–∞ –∏ –≤–æ–∑–º–æ–∂–Ω–æ –≤ –∏—Å—Ç–æ—á–Ω–∏–∫–µ —É–∫–∞–∑–∞–Ω –∏–º–µ–Ω–Ω–æ –∞–Ω–∞–ª–æ–≥–∏—á–Ω—ã–π —Ä–µ–π—Ç–∏–Ω–≥
</div>    


````python
for row in data.groupby('genre'):
    row_row = row[1]
    name_genre = row[0]
    
    median = (row_row['user_score'] / row_row['critic_score']).median()
   
    data.loc[~(data['user_score'].isna()) & (data['critic_score'].isna()) & (data['genre'] == name_genre), 'critic_score'] = \
    data.loc[~(data['user_score'].isna()) & (data['critic_score'].isna()) & (data['genre'] == name_genre), 'user_score'] / median
    
    data.loc[(data['user_score'].isna()) & ~(data['critic_score'].isna()) & (data['genre'] == name_genre), 'user_score'] = \
    data.loc[(data['user_score'].isna()) & ~(data['critic_score'].isna()) & (data['genre'] == name_genre), 'critic_score'] * median
````


### –í—ã–≤–æ–¥:
–†–µ–≥–∏–æ–Ω—ã EU –∏ NA —Å—Ö–æ–∂–∏ –≤ –ø—Ä–µ–¥–ø–æ—á—Ç–µ–Ω–∏—è—Ö –∫–∞–∫ –≤ –ø–ª–∞—Ç—Ñ–æ—Ä–º–∞—Ö, —Ç–∞–∫ –∏ –≤ –∂–∞–Ω—Ä–∞—Ö. </br>
–†–µ–≥–∏–æ–Ω JP –Ω–µ—Å–∫–æ–ª—å–∫–æ –æ—Ç–ª–∏—á–∞–µ—Ç—Å—è, –±–æ–ª—å—à–∞—è –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã 3DS (–≤ EU –∏ NA —ç—Ç–æ–π –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã –Ω–µ—Ç –≤ —Ç–æ–ø-5), —Ç–∞–∫–∂–µ –ø–æ–ø—É–ª—è—Ä–Ω–æ—Å—Ç—å —É –∂–∞–Ω—Ä–∞ Role-Playing ( –≤ EU –Ω–µ –≤—Ö–æ–¥–∏—Ç –≤ —Ç–æ–ø-5, –≤ NA –Ω–∞ 5-–º –º–µ—Å—Ç–µ)

<a id='hypothesis'></a>
# –®–∞–≥ 5. –ü—Ä–æ–≤–µ—Ä—å—Ç–µ –≥–∏–ø–æ—Ç–µ–∑—ã

### –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ;

–î–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –≥–∏–ø–æ—Ç–µ–∑–∞: </br>
–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: H0: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. user_score_xbox == user_score_pc </br>
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –ø–ª–∞—Ç—Ñ–æ—Ä–º Xbox One –∏ PC –ù–ï –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. user_score_xbox != user_score_pc</br>
–í–æ–∑—å–º—É —É—Ä–æ–≤–µ–Ω—å —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–æ–π –∑–Ω–∞—á–∏–º–æ—Å—Ç–∏ alpha = 0.05

<div class="alert alert-success"; style="border-left: 7px solid green">
<h3>‚úî –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</h3>
    
–ì–∏–ø–æ—Ç–µ–∑—ã —Å—Ñ–æ—Ä–º—É–ª–∏—Ä–æ–≤–∞–Ω—ã –≤–µ—Ä–Ω–æ üëç</div>   

–ü–æ—Å–º–æ—Ç—Ä—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –Ω–∞ XBox One –∏ PC, —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å –∫–∞–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä –±—Ä–∞—Ç—å equal_var = True –∏–ª–∏ equal_var = False –¥–ª—è –º–µ—Ç–æ–¥–∞ scipy.stats.ttest_ind (array1, array2, equal_var)

````python
user_score_xbox = data_actual.loc[(data['platform'] == "XOne") & ~(data_actual['user_score'].isna()),'user_score']
user_score_pc   = data_actual.loc[(data['platform'] == "PC") & ~(data_actual['user_score'].isna()),'user_score']
alpha = 0.05
````

–¢.–∫. –≤—ã–±–æ—Ä–∫–∏ —Ä–∞–∑–Ω–æ–≥–æ —Ä–∞–∑–º–µ—Ä–∞, equal_var = False.

<div class="alert alert-success"; style="border-left: 7px solid green">
<h3>‚úî –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞</h3>

–î–≤—É—Å—Ç–æ—Ä–æ–Ω–Ω—è—è –≥–∏–ø–æ—Ç–µ–∑–∞: </br>
–ù—É–ª–µ–≤–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: H0: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports –æ–¥–∏–Ω–∞–∫–æ–≤—ã–µ. action_score == sports_score</br>
–ê–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–∞—è –≥–∏–ø–æ—Ç–µ–∑–∞: –°—Ä–µ–¥–Ω–∏–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∏ –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports —Ä–∞–∑–Ω—ã–µ. action_score != sports_score </br>

–ü–æ—Å–º–æ—Ç—Ä—é –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏–≥—Ä –∂–∞–Ω—Ä–æ–≤ Action –∏ Sports, —á—Ç–æ–±—ã –ø–æ–Ω—è—Ç—å –∫–∞–∫–æ–π –ø–∞—Ä–∞–º–µ—Ç—Ä –±—Ä–∞—Ç—å equal_var = True –∏–ª–∏ equal_var = False –¥–ª—è –º–µ—Ç–æ–¥–∞ scipy.stats.ttest_ind (array1, array2, equal_var)

````python
action_score = data_actual.loc[(data['genre'] == "Action") & ~(data['user_score'].isna()),'user_score']
sports_score = data_actual.loc[(data['genre'] == "Sports") & ~(data['user_score'].isna()),'user_score']
alpha = 0.05
````

````python
results = st.ttest_ind (action_score, sports_score, equal_var=False)

print('p-–∑–Ω–∞—á–µ–Ω–∏–µ:', results.pvalue)

if results.pvalue < alpha:
    print("–û—Ç–≤–µ—Ä–≥–∞–µ–º –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
else:
    print("–ù–µ –ø–æ–ª—É—á–∏–ª–æ—Å—å –æ—Ç–≤–µ—Ä–≥–Ω—É—Ç—å –Ω—É–ª–µ–≤—É—é –≥–∏–ø–æ—Ç–µ–∑—É")
````
